```python

model_choice = 'A'
model_version = "Gilgamesh"
model_type = "Conditional"

resource_path = "/scratch/Jack/resources/Attention Maps/view maps/"
result_path =  "/scratch/Jack/research lab/NFL Viewpoint VAEs/"
cVAE = vr.VAE_Adventures(log=True, model_version=model_version, model_choice=model_choice, model_type=model_type, resource_path=resource_path, result_path=result_path)

epochs = 1000
for epoch in range(epochs):
    for i, datum in enumerate(v.trainloader_prediction):

        # Get z's for entire sequence
        in_datum_batch = vr.Variable(datum[1].cuda(), volatile=True)
        Z, z_mu, z_var, out_datum = v.VAE(in_datum_batch)

        tensor_conditional = None
        # Randomly pick a z pair
        pair_list = []
        while tensor_conditional == None or tensor_conditional[0].size()[0] < cVAE.batch_size:
            z_choice = vr.random.randint(1,datum[1].size()[0])
            z_window = [0,min(z_choice+cVAE.window_context[0], datum[1].size()[0])]
            if cVAE.window_context[1]:
                z_window[0] = max(0,z_choice-cVAE.window_context[0])
            possible_vals = [x for x in range(z_window[0], z_window[1]+1) if x != z_choice]
            random_pair = vr.random.choice(possible_vals)
            Zc = Z[z_choice-1].unsqueeze(0)
            Zn = Z[random_pair-1].unsqueeze(0)
            Zplus = torch.cat([Zc, Zn],1)
            if tensor_conditional is None:
                tensor_conditional = [Zc,Zn , Zplus ]
            else:
                tensor_conditional = [vr.torch.cat([tensor_conditional[0],Zc]), vr.torch.cat([tensor_conditional[1],Zn]), vr.torch.cat([tensor_conditional[2],Zplus])]

        myloss = cVAE.steptrain(tensor_conditional[0].data, tensor_conditional[1].data, tensor_conditional[2].data)
        print("Loss is : {}".format(myloss.numpy()[0]))
    
    
    
```

```python
### Which model to visualize, sets include {[4,64,512]} and epochs in [100,200,...,1000]

model_choice = 'A'
model_version = "Gilgamesh"
model_type = "Vanilla"
resource_path = "/scratch/Jack/resources/Attention Maps/view maps/"
result_path =  "/scratch/Jack/research lab/NFL Viewpoint VAEs/"

z_sub = 12 # how many z dimensions are interactive

### ------------------------------------------------------------------------------------

z_sub = min(4,z_sub)
v = vr.VAE_Adventures(log=False, model_version=model_version, model_choice=model_choice, model_type=model_type, resource_path=resource_path, result_path=result_path)
v.load_model()

### ------------------------------------------------------------------------------------


def visualize_z_sample(**z):
    sample_z = torch.zeros([1,z_sub]).cuda()
    sample_z_condition = torch.zeros([1,z_sub]).cuda()
    for i, (arg, val) in enumerate(z.items()):
        if arg != "wimage":
            if i < z_sub:
                sample_z[0,i] = val
            else:
                sample_z_condition[0,i-z_sub] = val

    v.inference = True
    sample_var = vr.Variable(sample_z).cuda()
    sample_var_condition = vr.Variable(sample_z_condition).cuda()
    print(sample_var, sample_var_condition)
    sample_output = cVAE.VAE.decode(sample_var, sample_var_condition)
    #sample_output = vr.F.sigmoid(sample_output)
    print(sample_output)
    #sample = v.VAE.decode(sample_var)
    
    to_pil = vr.transforms.ToPILImage()
    sample_cpu = sample_output.data.cpu()
    vr.oracle.visualize_tensors([sample_cpu],file="./temp_image")
    wimage.value = open("./temp_image.png", "rb").read()
    display(wimage)
    

def widget_slider(name):
    mystyle = {'handle_color': 'white'}
    w = vr.widgets.FloatSlider(
        value=0,
        min=-3.0,
        max=3.0,
        step=0.1,
        disabled=False,
        continuous_update=False,
        orientation='vertical',
        readout=True,
        style=mystyle)
    
    return w

z_interactive = {'z-' + str(k): widget_slider('z-' + str(k)) for k in range(z_sub*2)}

wimage =vr.widgets.Image(
    format='png',
    width=512,
    height=512,
)

hlay = vr.Layout(display='flex',
                    flex_flow='row',
                    align_items='stretch',
                    border='solid',
                    width='50%')

w = vr.interactive(visualize_z_sample, **z_interactive)
w.layout = hlay

### This acts as a predictor for the z-mu values that produce every frame

# for i, datum in enumerate(v.trainloader_prediction):
#     orig_datum = Variable(datum[0])
#     in_datum = Variable(datum[1].cuda())
#     v.inference = True
#     Z, z_mu, z_var, out_datum = v.VAE(in_datum)
#     print(Z)

#   Validation set

# validate = True
#
# if validate:
#     for i, datum in enumerate(v.testloader_prediction):
#
#         orig_datum = Variable(datum[0])
#         in_datum = Variable(datum[1].cuda())
#         v.inference = True
#         Z, z_mu, z_var, out_datum = v.VAE(in_datum)
#         loss = v.loss_function(out_datum, in_datum,  z_mu, z_var)
#         print(Z, loss)
        
display(w)
```
